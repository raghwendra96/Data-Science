```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "svg"
```

```python
raw_df = pd.read_csv("Zomato_Mumbai_Dataset.csv",delimiter='|')
```

```python
raw_df.head()
```

```python
raw_df.shape
```

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
# Checking redundunt cells of dataset
wrong_data = raw_df['PAGE NO'] == 'PAGE NO'
raw_df[wrong_data]
```

```python
## Performing Negation of the wrong dataset and then storing the correct data back in the raw_df DataFrame
## This permamnently remove the wrong data from the original dataframe
raw_df = raw_df[~wrong_data]
```

```python
# Dropping columns which are not required for further analysis
raw_df.drop(['URL', 'PAGE NO', 'CITY'], axis = 1, inplace=True)
```

```python
raw_df.head()
```

```python
# Checking for Null records
raw_df.isnull().sum()
```

```python
# Checking for a null row
raw_df[raw_df['PRICE'].isnull()]
```

```python
# Droping the above row from the dataset
raw_df = raw_df.drop(labels=15080, axis=0)
```

```python
# Replacing the other null records with NA
raw_df.fillna('NA', inplace=True)
```

```python
# Confirming all the null records are correct
raw_df.isnull().sum()
```

```python
# Checking for text values in the column before converting it to numeric datatype
raw_df['RATING'].value_counts()
```

```python
# Replacing the text values with '0'
raw_df['RATING'].replace(to_replace=['-','NEW','Opening'], value='0', inplace=True)
```

```python
# Checking for text values in the column before converting it to numeric datatype
raw_df['VOTES'].value_counts()
```

```python
# Replacing the text values with '0'
raw_df['VOTES'].replace(to_replace=['-','NEW','Opening'], value='0', inplace=True)
```

```python
# Changing Data Type of the numerical columns
raw_df['PRICE'] = raw_df.PRICE.astype('int64')
raw_df['RATING'] = raw_df.RATING.astype('float64')
raw_df['VOTES'] = raw_df.VOTES.astype('int64')
```

```python
raw_df.info()
```

```python
raw_df['TIMING'].value_counts()
```

```python
# Splitting the column and storing it in temp_df dataframe
temp_df = raw_df['TIMING'].str.split("(", n = 1, expand = True)
temp_df
```

```python
# Assigning the columns back to the raw_df dataframe
raw_df['TIMING'] = temp_df[0]
raw_df['DAYS_OPEN'] = temp_df[1]
raw_df.head()
```

```python
# Removing the bracket character from Days column
raw_df['DAYS_OPEN'] = raw_df['DAYS_OPEN'].str.replace(')','',regex=True)
raw_df.head()
```

```python
# Checking for Null records in DAYS_OPEN column
raw_df.isnull().sum()
```

```python
# Replacing the Null values with 'NA'
raw_df.fillna('NA', inplace=True)
```

```python
# Checking info of all the columns
raw_df.info()
```

```python
#finding data whose rating is 0 or may be vote is 0
useless_data = (raw_df['RATING'] == 0.0) | (raw_df['VOTES'] == 0)
raw_df[useless_data]
```

```python
#delete the useless data and restore the useful data in the current dataframe
raw_df = raw_df[~useless_data]
```

```python
#check value in RATING_TYPE which uniquely define
raw_df['RATING_TYPE'].value_counts()
```

```python
# translate other languages rating in standard english 
raw_df['RATING_TYPE'].replace(to_replace='Excelente' , value='Excellent' , inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Velmi dobré','Bardzo dobrze','Muy Bueno','Veľmi dobré'] ,value='Very Good', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Skvělá volba','Dobrze','Buono','Bueno','Dobré','Bom','Skvělé'] ,value='Good', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Çok iyi','Média','Priemer'], value='Average' , inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Průměr','Promedio','Ortalama','Muito Bom','İyi'], value='Poor', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Baik','Biasa','Media','Sangat Baik'] ,value='Very Poor', inplace=True)
```

```python
#check all the values
raw_df['RATING_TYPE'].value_counts()
```

```python
#working on the REGION column
raw_df['REGION'].value_counts()
```

```python
#remove unnecessary content which associated with locality 
raw_df['REGION']=raw_df['REGION'].str.replace('[a-zA-Z].+--' , '' , regex=True)
```

```python
#remove east and west terms from dataset
raw_df['REGION']=raw_df['REGION'].str.replace('West|west|East|east','' , regex=True)
```

```python
raw_df['REGION'].value_counts()
```

```python
#replace small region with known places
raw_df['REGION'] = raw_df['REGION'].str.replace('4 Bungalows|Andheri|7 Andheri|Azad Nagar|Near Andheri  Station| Near Andheri Station|Veera Desai Area','Andheri', regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Bandra Kurla Complex','Bandra',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('CBD-Belapur','CBD Belapur',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Girgaon Chowpatty','Chowpatty',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Dadar Shivaji Park','Dadar',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Flea Bazaar Café|Kamala Mills Compound','Lower Parel',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Runwal Green','Mulund',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Mumbai CST Area','Mumbai Central',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Kopar Khairane|Seawoods|Turbhe|Ulwe','Navi Mumbai',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('New Panvel|Old Panvel','Panvel',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Kamothe','Sion',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Ghodbunder Road|Majiwada','Thane',regex=True)
```

```python
raw_df[raw_df.duplicated()]
```

```python
#drop duplicates rows in dataset
raw_df = raw_df.drop_duplicates()
```

```python
frame_df = raw_df.copy()
```

```python
frame_df.head()
```

```python
fig = px.histogram(frame_df , x='CUSINE TYPE' , color='CUSINE TYPE',
                  title='Number of Restaurants by Cuisine Types' , 
                  labels={'CUSINE TYPE':'cuisine type'})
fig.show()
```

```python
rating_df = frame_df['RATING_TYPE'].value_counts().reset_index()
rating_df.rename(columns={'index':'RATING TYPE' , 'RATING_TYPE':'Number of Restaurants'} , inplace=True)
rating_df
```

```python
fig = px.pie(rating_df , names='RATING TYPE' , values='Number of Restaurants' ,color = 'RATING TYPE' ,
            title = 'Percentage of Restaurants by their Ratings').update_traces(textposition = 'inside',textinfo='percent+label')
fig.show()
```

```python
#top 10 restaurants which serves North Indian food as their signature dish
north_indian = frame_df[frame_df['CUSINE_CATEGORY'].str.contains('North Indian')]
north_indian.sort_values(by='RATING' , ascending=False).head(10)
```

```python
#Restaurants with excellent rating for their food service and rated above 4.5
high_rated_df = frame_df[(frame_df['RATING_TYPE'] == 'Excellent') & (frame_df['RATING'] >= 4.5)]
high_rated_df
```

```python
fig = px.histogram(high_rated_df, x='REGION' , color='CUSINE TYPE',
                  title = 'High Rated Restaurants Along With Their Rating').update_xaxes(categoryorder='total descending')
fig.show()
```

```python
# find average price for the all the cusine in the city of Mumbai
avg_rated_df = high_rated_df.groupby(by=['REGION','CUSINE TYPE'])['PRICE'].mean().reset_index()
avg_rated_df.head()
```

```python
fig = px.scatter(avg_rated_df , x='REGION' , y='PRICE' , color='CUSINE TYPE' ,symbol='CUSINE TYPE' , 
                title ='Avg Price for All Cusine in Area of Mumbai').update_traces(marker_size=10)
fig.show()
```

```python
# which area has large number of north indian restaurant market 
north_indian_df = frame_df[frame_df['CUSINE_CATEGORY'].str.contains('North Indian')]
north_indian_df
```

```python
north_df = north_indian_df.groupby(by='REGION').agg({'NAME':'count' ,'PRICE':'mean'}).rename(columns={'NAME':'COUNT OF RESTAURANT' , 'PRICE':'PRICE'}).reset_index()
north_df = north_df.sort_values('COUNT OF RESTAURANT' , ascending = False).head(50)
north_df.head()
```

```python
fig = px.bar(north_df , x='REGION' , y='COUNT OF RESTAURANT', color='PRICE',
            title = 'Restaurants According To their Popularity in Area\'s')
fig.show()
```

```python
# Relationship between price and rating for the dataset
rating_price_df = frame_df.groupby(['CUSINE TYPE','RATING'])['PRICE'].mean().reset_index()
rating_price_df
```

```python
fig = px.line(rating_price_df , x='CUSINE TYPE' , y='PRICE' , color='RATING')
fig.show()
```

```python
fig = px.line(rating_price_df , x='RATING' , y='PRICE' , color='CUSINE TYPE')
fig.show()
```

```python
#relationship between region and price
region_price_df = frame_df.groupby(['REGION'])['PRICE'].mean().drop_duplicates().reset_index()
region_price_df
```

```python
fig = px.line(region_price_df , x='REGION' , y='PRICE')
fig.show()
```
